##   matplotlib

```python
import matplotlib.pyplot as plt

fig = plt.figure()
# fig.set(facecolor='green')
fig.set_facecolor('green')
ax = fig.add_subplot(111)
ax.set(facecolor='orange')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

fig.set(facecolor='green')
ax.set(facecolor='red',
       xlim = [-10, 10],
       ylim = [-2, 2],
       title = 'Title',
       xlabel = 'labelX', 
       ylabel = 'labelY')

plt.show()
```
